### Employee Turnover

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score
```

```python
emp_data = pd.read_csv("employee_turnover.csv")
emp_data.head()
```

```python
emp_data.info()
```

```python
emp_data.describe()
```

```python
X = emp_data.drop('Employee_Turnover', axis=1)
y = emp_data['Employee_Turnover']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(y_train[y_train == 1].count())
print(y_train[y_train == 0].count())
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
```

```python
# L2 Regularization using Ridge Classifier
from sklearn.linear_model import RidgeClassifier
ridge_model = RidgeClassifier()

ridge_model.fit(X_train, y_train)
ridge_y_pred = ridge_model.predict(X_test)

ridge_accuracy = accuracy_score(y_test, ridge_y_pred)
ridge_precision = precision_score(y_test, ridge_y_pred)

print(f"Ridge Accuracy: {ridge_accuracy}")
print(f"Ridge Precision: {ridge_precision}")    
```

```python
# L1 Regulerization - there is no Lasso Classifier in sklearn, but we can use Logistic Regression with L1 penalty

l1_model = LogisticRegression(
    penalty='l1',
    solver='liblinear',   # or solver='saga'
    random_state=42
)

l1_model.fit(X_train, y_train)

l1_y_pred = l1_model.predict(X_test)

l1_accuracy = accuracy_score(y_test, l1_y_pred)
l1_precision = precision_score(y_test, l1_y_pred)

print(f"L1 Logistic Accuracy: {l1_accuracy}")
print(f"L1 Logistic Precision: {l1_precision}")

# The accuracy and precision are increased slightly
```
