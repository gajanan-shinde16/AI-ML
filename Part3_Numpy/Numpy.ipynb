## Lists vs Np array

### Lets see time/performance

```python
import numpy as np
import time
```

```python
size = 1_000_0000

py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()


print(f"time taken by List = {end-start}")

np_arr = np.array(py_list)
start = time.time()
sq_arr = np_arr**2
end = time.time()


print(f"time taken by np Array = {end-start}")
```

### lets see memory occupied

```python
import sys
print(f"Size occupied by Python List = {sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"Size occupied by np Array = {np_arr.nbytes} bytes")
```

## Important Operations on Arrays

### 1. reshape

```python
arr = np.array([[1,2,3],[4,5,6]])

arr2 = arr.reshape((3,2))
arr3 = arr.flatten();

print(arr)
print(arr2)
print(arr3)
```

### 2. Indexing

```python
arr = np.array([1,2,3,4,5])
print(arr[1])

arr[3] = 10
print(arr)

arr = np.array([[1,2,3],[4,5,6]])
print(arr[0][2])
```

```python
# fancy indexing
arr = np.array([1,2,3,4,5])
idx = [1,3,4]
print(arr[idx])# only those elements returned whose index is passed
```

```python
#boolean indexing
arr = np.array([1,2,3,4,5])
print(arr[arr>3])
print(arr[arr%2==0])
```

### 3. Slicing

slicing in numpy array is similar to python lists but there is a big difference ij how internally slicing takes place

- py list slicing -> creates a new copy of elements/subarray
- np array creates a view of original array which still points to subarray/subsequence within orig array
  to verify just update one elements in both and notice the change

```python
arr = np.array([1,2,3,4,5])

print(arr[1:3])#[start,end,step]
print(arr[:3])
print(arr[1:])
print(arr[::2]) # alternate
```

## Vectorization in NumPy

Vectorization means replacing Python loops with fast NumPy operations.

```python
#without vectorization
a = [1, 2, 3, 4]
b = []

for x in a:
    b.append(x * 2)

print(b)
```

```python
# with vectorization (Numpy)

a = np.array([1, 2, 3, 4])
b = a * 2
print(b)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x + y)      # element-wise addition
print(x ** 2)     # element-wise power
print(np.sin(x))  # vectorized sin()
```

## Broadcasting in NumPy

Broadcasting allows NumPy to operate on arrays with different shapes by automatically expanding dimensions when possible.

###  Broadcasting Rules (simple version)

Two arrays can be broadcast if for each dimension:

They are equal, OR

One of them is 1

NumPy then stretches the dimension with value 1 to match the other.

```python
arr = np.array([1, 2, 3])
print(arr + 5)
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(A + b)

# A.shape = (2, 3)

# b.shape = (3,) → broadcast to (1, 3) → (2, 3)
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([[10],
              [20]])

print(A + b)
```

## Mathematical Functions (some mostly used)

```python
arr = np.array([5, 10, 15, 20, 25])

# ---- Aggregation Functions ----  -->perform ops on all and give single result
print("sum:", arr.sum())          # total of all elements
print("prod:", arr.prod())        # product of all elements
print("min:", arr.min())          # smallest value
print("max:", arr.max())          # largest value
print("mean:", arr.mean())        # average
print("std:", arr.std())          # standard deviation
print("var:", arr.var())          # variance
print("median:", np.median(arr))  # middle value
print("argmin:", arr.argmin())    # index of min value
print("argmax:", arr.argmax())    # index of max value
```

```python
# Power Functions

arr = np.array([1, 4, 9, 16])

print("square:", np.square(arr))      # x²
print("sqrt:", np.sqrt(arr))          # √x
print("pow:", np.power(arr, 3))       # x³
print("pow with 2 args:", np.power(2, 5))  # 2⁵
```

```python
# Rounding functions 
arr = np.array([1.2, 2.7, -3.8, 4.5])

print("round:", np.round(arr))   # nearest integer
print("floor:", np.floor(arr))   # always down
print("ceil:", np.ceil(arr))     # always up
print("trunc:", np.trunc(arr))   # remove decimal part
```

```python
# Log & Exponential
arr = np.array([1, 10, 100, 1000])

print("log (natural):", np.log(arr))     # ln(x)
print("log10:", np.log10(arr))           # base-10
print("log2:", np.log2(arr))  
arr = np.array([0, 1, 2, 3])# base-2
print("exp:", np.exp(arr))               # e^x
```

```python
print("unique:", np.unique(arr))   # remove duplicates
print("sort:", np.sort(arr))       # ascending order
print("abs:", np.abs([-3, 0, 7]))  # absolute value
```
