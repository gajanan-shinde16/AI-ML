```python
# np.random.seed(0)

import numpy as np

np.random.seed(0)

print(np.random.random(3))
np.random.seed(0)
# set again to 0 so exact same op
print(np.random.random(3))
# here it will change
print(np.random.random(3))
```

```python
np.random.seed(0)
print(np.random.randint(10))
print(np.random.randint(10))

np.random.seed(0)
print(np.random.randint(10))
np.random.seed(0)
print(np.random.randint(10))

print(np.random.random(5)) # b/n 0 and 1
print(np.random.randint(10,20,5)) # (low,high=None,size=None)
print(np.random.rand(2,5))
print(np.random.choice([10,20,30]))
print(np.random.choice([10,20,30]))

arr = np.array([1,2,3,4,5])
print(arr)
np.random.shuffle(arr)
print(arr)

arr2 = np.array([5,6,7,8,9])
copy = np.random.permutation(arr2)
print(arr2)
print(copy)

print(np.random.uniform(0,10,5))
print(np.random.normal(0,10,5))
print(np.random.binomial(10,0.5,5))
print(np.random.poisson(5,10))






```
