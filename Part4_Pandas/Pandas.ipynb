## Data Transformation (feature Engineering )


```python
# 1 apply()

df2 = df.copy()
df2["tax"] = df["income"].apply(lambda x : "20%" if x>60000 else "10%")
df2
```

```python
# map()

df2 = df.copy()
gender_map = {"Male" : "M" , "Female" : "F" , "Unknown" : "U"}

df2["gender"] = df2["gender"].map(gender_map)
df2
```

```python
# 3 assign()

df2 = df.copy()
df2 = df2.assign(new_income = df2["income"] * 1.1) # 10% inc
df2
```

```python
# 4 replace(old,new)

df2 = df.copy()
df2["country"] = df2["country"].replace("USA" , "United States")
df2
```

```python
# 5 rename
df2.columns = ["Id" , "Name", "Age" , "Country" , "Gender" , "Income"] # make cols capital

df2 = df2.rename(columns = {"Income" : "Salary"})
df2 = df2.rename(index = {1 : "First"})
df2
```

## Melt and Pivot

```python
df = pd.DataFrame({
    "country" : ["USA" ,"USA" ,"India", "India"],
    "year" : [2020,2021,2020,2021],
    "sales" : [100,120,90,110],
    "profit" : [20,25,18,22],
    "tax" : [10,15,20,28]
})

df
```

### MELT

```python
melted_df = df.melt(
    id_vars = ["country" , "year"] , 
    value_vars = ["sales" ,"profit", "tax"] , 
    var_name = "matrics" , 
    value_name = "value"
)
```

```python
melted_df
```

### Pivot

```python
orig = melted_df.pivot(
    index = ["country", "year"], # index should be unique so used two attributes
    columns = "matrics",
    values="value"
)
```

## Merge and join

```python
customer_df = pd.DataFrame({
    "customer_id" : [1,2,3,4],
    "name" : ["Gajanan", "Pratik", "Prasad", "Pramod"]
})

order_df = pd.DataFrame({
    "order_id" : [101,102,104,105],
    "customer_id" : [2,1,3,5],
    "amount" : [102,234,245,654]
})


```

```python
customer_df
```

```python
order_df
```

```python
pd.merge(customer_df,order_df , on = "customer_id") # inner join -->only commen parameter values will be displayed
```

```python
pd.merge(customer_df,order_df , on = "customer_id" ,how="left") #left join-->means all values from left table will  included uses NaN for unvailable orders
```

```python
pd.merge(customer_df,order_df , on = "customer_id" ,how="right") #right join-->means all values from right table will  included uses NaN for unvailable orders
```

```python
pd.merge(customer_df,order_df , on = "customer_id" ,how="outer") #outer join-->means all values from left table nd right table will  included uses NaN for unvailable orders
```
