## Data load

```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
print(data)
print(type(data))
```

## Clean DATA
### data is inconsistent--ratings types,missing values like age

```python
def clean_data(data):
    text_to_num = {"one":"1","two":"2","three":"3","four":"4","five":"5"}
    clean_data = []
    unique_users = set()

    
    for user in data:
        #clean ratings -data consistency
        raw_rating = user["rating"].strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # handle missing values -

        raw_age = user.get("age")
        if(raw_age==None):
            user["age"] = None

        # handle duplicasy
        if user["name"].strip() in unique_users:
            continue
        clean_data.append(user)
        unique_users.add(user["name"].strip())
    return clean_data
```

```python
data = clean_data(data)
print(data)
```

## Getting meaningful insights 

```python
def get_insights(data):

    #get avg rating
    tot_rating=0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"Average Rating: {tot_rating/len(data)}")
    poor_ratings=0
    for user in data:
        if(float(user["rating"])<3):
            poor_ratings += 1
    print(f"% of poor Rating: {poor_ratings/len(data)*100}%")
```

```python
get_insights(data)
```
