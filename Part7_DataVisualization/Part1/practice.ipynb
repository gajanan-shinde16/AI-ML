### Problem 6

#### Compare two bars

```python
import numpy as np

products = ["Laptop", "Mobile", "Tablet", "Monitor", "Keyboard"]
company_a = [50, 70, 40, 60, 45]
company_b = [55, 65, 35, 70, 50]

w = 0.35
x = np.arange(len(products))

plt.bar(x - w/2, company_a, width=w, color='blue', label='Company A')
plt.bar(x + w/2, company_b, width=w, color='orange', label='Company B')

plt.xticks(x, products)      # Show product names
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Company Sales Comparison")
plt.legend()
plt.ylim(0,max(company_a) + 20 )
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Bar Plot Practice Problem
Tasks   
1. Create a bar chart.
2. Add the title "Python Test Scores".
3. Label both axes.
4. Set the bar width to 0.5.
5. Give each bar a different color.
6. Display the exact score above each bar.
7. Rotate x-axis labels by 30°.
8. Add horizontal grid lines only.
9. Highlight the highest score using a different edge color and thicker border.
10. Save the figure as scores.png.

```python
# practice
import matplotlib.pyplot as plt

students = ["Aman", "Riya", "Karan", "Neha", "Vikas"]
marks = [78, 92, 65, 88, 95]
max_mark = max(marks)
edgecolors = ['black' if mark == max_mark else 'gray' for mark in marks]
linewidths = [2.5 if mark == max_mark else 1 for mark in marks]
colors = ['purple','green','blue','violet','yellow']

plt.bar(students,marks,width=0.5,color=colors,edgecolor=edgecolors,linewidth=linewidths)
plt.title("Python Test Scores")
plt.xlabel('Students')
plt.ylabel("Marks")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(10)
for i in range(len(students)):
    plt.text(students[i],marks[i]+2,str(marks[i]),ha='center')

plt.savefig("scores.png")
plt.show()
```

### Line Plot Practice Problem

Task    
1. Plot the data using a line plot.
2. Add a title: "Daily Website Visitors"
3. Label the x-axis
4. Label the y-axis
5. Use circle ( o ) markers.
6. Display grid lines.
7. Change the line width to 3
8. Color the line green.
9. Annotate the highest visitor count on the graph.
10. Save the figure as "visitors.png"


```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
visitors = [120, 150, 170, 160, 210, 260, 240]

max_visitors = max(visitors)

plt.plot(days, visitors, marker='o', linewidth=3, color='green')
plt.title("Daily Website Visitors")
plt.xlabel("Days")
plt.ylabel("Visitors")
plt.ylim(0,max_visitors+40)

# for day,vis in zip(days,visitors):
#     if vis == max_visitors:
#         plt.text(day,vis+5,str(vis),ha='center')

for day, vis in zip(days, visitors):
    if vis == max_visitors:
        plt.annotate(
            f"Peak: {max_visitors}",
            xy=(day, vis), # position
            xytext=(day, vis + 20),# Position of the label
            ha='center',
            arrowprops=dict(arrowstyle="->",color='red'),
            color='purple'
        )

plt.grid()
plt.show()


```
