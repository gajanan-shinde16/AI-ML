#### Multiple datasets on a scatter plot

```python
cities = ['Pune','Mumbai','Nagpur','PCMC','Nanded','Nashik','Parbhani','Thane', 'Hingoli', 'Jalana' ]

winter_temp = [3, 5, 2, 0, 4, 6, 1, 7, 5, 2]
winter_humidity = [82, 78, 85, 88, 80, 75, 90, 73, 79, 84]

summer_temp = [26, 29, 31, 34, 36, 38, 33, 30, 35, 37]
summer_humidity = [65, 60, 58, 55, 52, 50, 57, 62, 54, 51]

# plt.scatter(winter_temp, winter_humidity, label='Winter')
# plt.scatter(summer_temp, summer_humidity, label='Summer')

plt.scatter(winter_temp, winter_humidity, label='Winter',color='purple',s=40)
plt.scatter(summer_temp, summer_humidity, label='Summer',color='red',s=60)
plt.xlabel("Temp (celc)")
plt.ylabel("Humidity (%)")
plt.legend()
for i in range(len(cities)):
    plt.annotate(
        cities[i],
        xy=(winter_temp[i],winter_humidity[i]),
        xytext=(winter_temp[i], winter_humidity[i]+2)

    )

    plt.annotate(
        cities[i],
        xy=(summer_temp[i],summer_humidity[i]),
        xytext=(summer_temp[i], summer_humidity[i] +2)
    )

plt.show()
```

### Pie Chart

```python
expenses = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
amount = [15000, 7000, 3000, 2000, 8000]

plt.pie(amount, labels=expenses)
plt.title("Monthly Expenses")
plt.show()
```

#### customizations 

```python
expenses = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
amount = [15000, 7000, 3000, 2000, 8000]
colors = plt.cm.viridis(np.linspace(0, 1, len(amount)))
explodes = [0,0,0,0,0.2]
plt.pie(
    amount,
    labels=expenses,
    autopct='%1.1f%%',
    colors=colors,
    wedgeprops={
        'edgecolor':'black',
        'linewidth':1,
        'linestyle': '--',
        # 'fill':False
    },
    explode=explodes,
    shadow=True
    )
plt.title("Monthly Expenses")
plt.show()
```

#### two/multiple datasets on histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
group_a = np.random.normal(170, 7, 500)
group_b = np.random.normal(178, 8, 500)

# Common bins
bins = np.linspace(
    min(group_a.min(), group_b.min()),
    max(group_a.max(), group_b.max()),
    25
)

plt.figure(figsize=(8, 5))


plt.hist(group_a, bins=bins, alpha=0.6, color='steelblue', label='Group A')
plt.hist(group_b, bins=bins, alpha=0.6, color='salmon', label='Group B')


mean_a = np.mean(group_a)
mean_b = np.mean(group_b)


plt.axvline(mean_a, color='blue', linestyle='--', linewidth=2, label=f'Group A Mean = {mean_a:.1f} cm')
plt.axvline(mean_b, color='red', linestyle='--', linewidth=2, label=f'Group B Mean = {mean_b:.1f} cm')

plt.title("Height Distribution Comparison")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

### Subplots

indexing is row-major

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10)
y1 = np.sqrt(x)
y2 = x*2
y3 = x**2
y4 = x**3

plt.subplot(2,2,1) # indexing starts with 1

plt.plot(x,y1)
plt.title("plot1- Square root")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("plot2- Double")

plt.subplot(2,2,3) 
plt.plot(x,y3)
plt.title("plot3- Squares")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("plot4- Cubes")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10)
y1 = np.sqrt(x)
y2 = x*2
y3 = x**2
y4 = x**3

plt.subplot(2,3,1) # indexing starts with 1

plt.plot(x,y1)
plt.title("plot1- Square root")

plt.subplot(2,3,2)
plt.plot(x,y2)
plt.title("plot2- Double")

plt.subplot(2,3,3) 
plt.plot(x,y3)
plt.title("plot3- Squares")

plt.subplot(2,3,4)
plt.plot(x,y4)
plt.title("plot4- Cubes")

plt.tight_layout()
plt.show()
```

## Modern Matplotlib
### object oriented API

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,y1)
ax[0][0].set_title("Square Root")
ax[0][0].set_xlabel("X values")
ax[0][0].set_ylabel("y values")

ax[0][1].plot(x,y2)
ax[0][1].set_title("Doubles")
ax[0][1].set_xlabel("X values")
ax[0][1].set_ylabel("y values")

ax[1][0].plot(x,y3)
ax[1][0].set_title("Squares")
ax[1][0].set_xlabel("X values")
ax[1][0].set_ylabel("y values")

ax[1][1].plot(x,y4)
ax[1][1].set_title("Cubes")
ax[1][1].set_xlabel("X values")
ax[1][1].set_ylabel("y values")

fig.suptitle("Multiple plots")
fig.tight_layout()
fig.show()
```
