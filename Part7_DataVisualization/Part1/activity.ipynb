Compare a temperature of cities over a week using subplots and loops

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Pune", "Mumbai"]

temperatures = [
    [22, 23, 21, 24, 25], # New York
    [18, 19, 17, 20, 21], # London
    [30, 32, 31, 33, 34], # Pune
    [25, 26, 24, 27, 28] # Mumbai

]

fig,axes = plt.subplots(2,2)
city_num = 0
for i in range(2):
    for j in range(2):
        city = cities[city_num]
        cityTemp = temperatures[city_num]

        axes[i][j].plot(days,cityTemp)
        axes[i][j].set_title(city)
        axes[i][j].set_xlabel("Days")
        axes[i][j].set_ylabel("Temperature")

        city_num += 1

fig.suptitle("Temperature of cities over a week")
fig.tight_layout()
plt.show()
```

by flattening an array

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Pune", "Mumbai"]

temperatures = [
    [22, 23, 21, 24, 25], # New York
    [18, 19, 17, 20, 21], # London
    [30, 32, 31, 33, 34], # Pune
    [25, 26, 24, 27, 28] # Mumbai

]

fig,axes = plt.subplots(2,2)
axes = axes.flatten()
for i in range(len(axes)):
        city = cities[i]
        cityTemp = temperatures[i]

        axes[i].plot(days,cityTemp)
        axes[i].set_title(city)

fig.suptitle("Temperature of cities over a week")
fig.supxlabel("Days")
fig.supylabel("Temperature")
fig.tight_layout()
plt.show()
```
