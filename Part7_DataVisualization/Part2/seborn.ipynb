# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

## Relplot (Relational plot)
sns.relplot()

### By def Scatter Plot

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.show()
```

### Line plot (kind='line')

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='line'
)
plt.show()
```

### multiple plots

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time"
)
plt.show()

# Doing the same with scatterplot() would require:
# fig, ax = plt.subplots(1, 2)
# sns.scatterplot(..., ax=ax[0])
# sns.scatterplot(..., ax=ax[1])
```

```python
sns.set_theme()
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker'
)
plt.show()
```

```python
sns.set_theme()
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size'
)
plt.show()
```

```python
sns.set_theme()
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size',
    style='smoker'
)
plt.show()
```

```python
x_val = [1,2,3,4,5,6,7,8,9]
y_val = [i**2 for i in x_val]
sns.relplot(
    x=x_val,
    y=y_val,
    kind='line'
)
plt.show()
```

```python
flight_data = sns.load_dataset('flights')
print(flight_data.head())

sns.lineplot(
    data=flight_data,
    x='year',
    y='passengers'
)
```

### in above plot the shading is  
Seaborn computes the mean value at each x and draws:  

Line → estimated mean  
Shaded area → confidence interval around that mean  
By default, Seaborn shows a 95% confidence interval.

mean +- CI == shaded area

```python
flight_data = sns.load_dataset('flights')

sns.lineplot(
    data=flight_data,
    x='year',
    y='passengers',
    ci=None
)
plt.show()
```

```python
flight_data = sns.load_dataset('flights')

sns.lineplot(
    data=flight_data,
    x='year',
    y='passengers',
    errorbar=None
)
plt.show()
```

```python
flight_data = sns.load_dataset('flights')

sns.lineplot(
    data=flight_data,
    x='year',
    y='passengers',
    errorbar= ('ci',90) # we can change CI
)
plt.show()
```

## Best Practices to use Seaborn and matplotlib

- use seaborn for individual plots
- use matplotlib for custpmizations

```python
fig,ax = plt.subplots()
sns.lineplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    ax=ax,
    marker='o',
    errorbar=None
)
ax.set_title("Avg bill value for each day")
ax.set_xlabel("Day")
ax.set_ylabel("Bill Value")
fig.tight_layout()
plt.show()
```
