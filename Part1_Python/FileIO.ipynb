# File Handling

## Opening a File

Used to access a file.
Returns a file object.  
Syntax:  
open(filename, mode)

```python
f = open("data.txt", "r")

print(f)

f.close()
```

## File Modes  
r - Read  
w - Write (overwrite)  
a - Append  
x - Create new file  
r+ - Read and Write  
w+ - Write and Read  
a+ - Append and Read  
rb - Binary Read  
wb - Binary Write  

```python
f = open("demo.txt", "w")
f.close()
```

## Closing Files  
Releases system resources    
Prevents data corruption

```python
f = open("data.txt", "r")
# operations
f.close()
```

## Using with Statement

Automatically closes file  
Recommended approach  
Cleaner syntax

```python
with open("data.txt", "r") as f:
    data = f.read()

print(data)
```

## write()
Writes a string  
Overwrites existing content in w mode 


```python
with open("file.txt", "w") as f:
    f.write("Hello Python")
```

## writelines()
Writes multiple strings  
Takes a list.

```python
lines = [
    "Apple\n",
    "Banana\n",
    "Mango\n"
]

with open("fruits.txt", "w") as f:
    f.writelines(lines)
```

## read()
Reads entire file  
Returns string

```python
with open("file.txt", "r") as f:
    content = f.read()
print(content)
```

## read(size)
Important Notes  
Reads specific number of characters

```python
with open("file.txt", "r") as f:
    data = f.read(5)
print(data)
```

## readline()
Reads only one line.

```python
with open("file.txt", "r") as f:
    line = f.readline()

print(line)
```

## readlines()
Reads all lines  
Returns list

```python
with open("file.txt", "r") as f:
    lines = f.readlines()
print(lines)
```

## Append Mode
Adds data at the end  
Existing content remains safe

```python
with open("file.txt", "a") as f:
    f.write("\nData Science")
```

## tell()
Returns cursor position

```python
with open("file.txt", "r") as f:
    print(f.tell())
    f.read(4)
    print(f.tell())
```

## seek()
Moves cursor to specified position

```python
with open("file.txt", "r") as f:
    f.seek(5)
    print(f.read())
```

## Checking File Existence
Requires os module

```python
import os
print(os.path.exists("file.txt"))
print(os.path.exists("file1.txt"))
```

## Creating a New File
Uses x mode  
Error occurs if file already exists

```python
with open("new.txt", "x") as f:
    f.write("Created")
```

## Rename File
Uses os.rename()

```python
import os
os.rename(
    "old.txt",
    "new.txt"
)
```

## Delete File
Uses os.remove()

```python
import os
os.remove("new.txt")
```

## Exception Handling
Avoids program crash  
Common exception:  
FileNotFoundError

```python
try:

    with open("abc.txt", "r") as f:
        print(f.read())

except FileNotFoundError:

    print("File not found")
```

## Binary Files
Used for:  
- Images
- Videos
- PDFs
- Audio files

```python
with open("data.bin", "wb") as f:
    f.write(b"Python")
```

```python
with open("data.bin", "rb") as f:
    print(f.read())
```
