# File Handling

## Opening a File

Used to access a file.
Returns a file object.  
Syntax:  
open(filename, mode)

```python
f = open("data.txt", "r")

print(f)

f.close()
```

## File Modes  
r - Read  
w - Write (overwrite)  
a - Append  
x - Create new file  
r+ - Read and Write  
w+ - Write and Read  
a+ - Append and Read  
rb - Binary Read  
wb - Binary Write  

```python
f = open("demo.txt", "w")
f.close()
```

## Closing Files  
Releases system resources    
Prevents data corruption

```python
f = open("data.txt", "r")
# operations
f.close()
```

## Using with Statement

Automatically closes file  
Recommended approach  
Cleaner syntax

```python
with open("data.txt", "r") as f:
    data = f.read()

print(data)
```

## write()
Writes a string  
Overwrites existing content in w mode 


```python
with open("file.txt", "w") as f:
    f.write("Hello Python")
```

## writelines()
Writes multiple strings  
Takes a list.

```python
lines = [
    "Apple\n",
    "Banana\n",
    "Mango\n"
]

with open("fruits.txt", "w") as f:
    f.writelines(lines)
```

## read()
Reads entire file
